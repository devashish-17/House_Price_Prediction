```python
def rangeToFloat(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1])) / 2
    try:
        return float(x)
    except:
        return None
```

```python
house['total_sqft'] = house['total_sqft'].apply(rangeToFloat)
house['total_sqft'].unique()
```

```python
house.head()
```

```python
house['price_per_sqft'] = house['price'] * 100000 / house['total_sqft']
house.head()
```

```python
house.describe()
```

```python
house['location'].value_counts()
```

```python
house['location'] = house['location'].apply(lambda x : x.strip())
location_count = house['location'].value_counts()
location_count
```

```python
location_count_less_10 = location_count[location_count <= 10]
location_count_less_10
```

```python
house['location'] = house['location'].apply(lambda x : 'other' if x in location_count_less_10 else x)
house['location'].value_counts()
```

## Outlier Detection and Removal

```python
house.describe()
```

```python
(house['total_sqft']/house['bhk']).describe()
```

```python
house = house[(house['total_sqft']/house['bhk']) >= 300]
house.describe()
```

```python
house.price_per_sqft.describe()
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        sd = np.std(subdf.price_per_sqft)
        gen_df = subdf[(subdf.price_per_sqft > (m-sd)) & (subdf.price_per_sqft <= (m+sd))]  # normal distribution (+ and - 1 Std Deviation) (65%)
        df_output = pd.concat([df_output, gen_df], ignore_index=True)
    return df_output
house = remove_outliers_sqft(house)
house.describe()
```

```python
def remove_outliers_bhk(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices, axis = 'index')
```

```python
house = remove_outliers_bhk(house)
house
```

# Model Evaluation

```python
X = house.drop(columns = ['price'])
Y = house['price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 0)
```

```python
print(X_train.shape, X_test.shape)
```

## Applying Liner Regression

```python
column_trans = ColumnTransformer(
    transformers=[
        ('location_encoder', OneHotEncoder(sparse_output=False), ['location']),
        ('passthrough_columns', 'passthrough', list(set(house.columns) - set(['location'])))
    ])

X_transformed = column_trans.fit_transform(house)

# Convert the NumPy array back to a DataFrame for better readability
X_df = pd.DataFrame(X_transformed, columns=column_trans.get_feature_names_out())

# print(house[X_df])
X_df
```

```python
scaler = StandardScaler()
```

```python
regressor = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, scaler, regressor)
```

```python
pipe.fit(X_train, Y_train)
```

```python
Y_pred_regressor = pipe.predict(X_test)
```

```python
r2_score(Y_test, Y_pred_regressor)
```

## Applying Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans, scaler, lasso)
```

```python
pipe.fit(X_train, Y_train)
```

```python
Y_pred_lasso = pipe.predict(X_test)
r2_score(Y_test, Y_pred_lasso)
```

## Applying Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans, scaler, ridge)
```

```python
pipe.fit(X_train, Y_train)
```

```python
Y_pred_ridge = pipe.predict(X_test)
r2_score(Y_test, Y_pred_ridge)
```

```python
print('Linear : ', r2_score(Y_test, Y_pred_regressor))
print('Lasso : ', r2_score(Y_test, Y_pred_lasso))
print('Ridge : ', r2_score(Y_test, Y_pred_ridge))
```
